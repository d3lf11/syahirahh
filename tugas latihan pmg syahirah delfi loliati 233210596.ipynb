```python
# perkalian matriks array

import numpy as np

a = np.array([[[2,6], [9,3],[7,5]]])
b = np.array([[[8,4], [4,2], [1,3]]])

kali = a*b
print(kali)
```

```python
# pengurangan matriks array

import numpy as np

a = np.array([[[8,6], [9,3],[7,5]]])
b = np.array([[[6,4], [4,2], [1,3]]])

kurang = a-b
print(kurang)
```

```python
# transpose matriks array

import numpy as np

a = np.array([[8,6,7], [9,3,4], [5,9,8]])
b = np.array([[6,4,7], [4,2,5], [1,3,3]])

transpose_a = np.transpose (a)
transpose_b = np.transpose (b)
print (transpose_a)
print ( transpose_b)
```

```python
# inverse matriks array

import numpy as np

a = np.array([[1,2], [9,3]])
b = np.array([[7,4], [ 6,3]])

inverse_a = np.linalg.inv(a)
inverse_b = np.linalg.inv(b)
print("\nInverse matrkis a:")
print(inverse_a)
print("\nInverse matrkis b:")
print(inverse_b)

```

```python
# determinan matriks array

import numpy as np

a = np.array([[8,2], [6,9]])
b = np.array([[2,5], [ 4,2]])

determinan_a = np.linalg.det(a)
determinan_b = np.linalg.det(b)
print("Determinan dari matriks a:", determinan_a) 
print("Determinan dari matriks b:", determinan_b)
```
